```python
#Problem - 4: Matrix Operations
import numpy as np
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
#1
A_inv = np.linalg.inv(A)
result = A.dot(A_inv)
print(result)
#2
AB = A.dot(B)
BA = B.dot(A)
print("AB =\n", AB)
print("BA =\n", BA)
print("Are AB and BA equal? ->", np.array_equal(AB,BA))
#3
left = (A.dot(B)).T
right = B.T.dot(A.T)
print("Left side =\n", left)
print("Right side =\n", right)
```

```python
#• Solve the following system of Linear equation using Inverse Methods.
# 2x − 3y + z = −1
# x − y + 2z = −3
# 3x + y − z = 9
import numpy as np
A = np.array([
    [ 2, -3,  1],
    [ 1, -1,  2],
    [ 3,  1, -1]
])
B = np.array([-1, -3, 9])
A_inv = np.linalg.inv(A)
X = A_inv.dot(B)
print(X)
```

```python
import numpy as np
import time

#ADDITION
# Python lists
list1 = list(range(1_000_000))
list2 = list(range(1_000_000))
start = time.time()
list_add = [list1[i] + list2[i] for i in range(1_000_000)]
end = time.time()
print("Python List Addition Time:", end - start, "seconds")
# NumPy arrays
arr1 = np.arange(1_000_000)
arr2 = np.arange(1_000_000)
start = time.time()
arr_add = arr1 + arr2
end = time.time()
print("NumPy Array Addition Time:", end - start, "seconds")

#MULTIPLICATION
# Python lists
start = time.time()
list_mul = [list1[i] * list2[i] for i in range(1_000_000)]
end = time.time()
print("Python List Multiplication Time:", end - start, "seconds")
# NumPy arrays
start = time.time()
arr_mul = arr1 * arr2
end = time.time()
print("NumPy Array Multiplication Time:", end - start, "seconds")

#DOT PRODUCT
# Python lists
start = time.time()
dot_py = sum(list1[i] * list2[i] for i in range(1_000_000))
end = time.time()
print("Python List Dot Product Time:", end - start, "seconds")
# NumPy arrays
start = time.time()
dot_np = np.dot(arr1, arr2)
end = time.time()
print("NumPy Dot Product Time:", end - start, "seconds")

#MATRIX MULTIPLICATION
# Python lists
M1 = [[i for i in range(1000)] for _ in range(1000)]
M2 = [[i for i in range(1000)] for _ in range(1000)]
start = time.time()
# matrix multiplication using python
result_py = [[sum(M1[i][k] * M2[k][j] for k in range(1000))
              for j in range(1000)]
              for i in range(1000)]
end = time.time()
print("Python List Matrix Multiplication Time:", end - start, "seconds")

# NumPy arrays
A = np.arange(1_000_000).reshape(1000, 1000)
B = np.arange(1_000_000).reshape(1000, 1000)
start = time.time()
result_np = np.dot(A, B)
end = time.time()
print("NumPy Matrix Multiplication Time:", end - start, "seconds")
```
